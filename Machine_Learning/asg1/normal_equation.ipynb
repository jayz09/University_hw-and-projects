TASK 2 Find value of the parameters of linear regression of model y = ax^2 + bx + c using the normal equation approach

```python
# Import Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (12.0,9.0)
```

```python
data = pd.read_csv('data_hw1.csv', header = 0, names = ['x_val', 'y_val'])
x = data['x_val']
y = data['y_val']
```

```python
# Initialization
x1 = x.to_numpy().reshape((len(x), 1)) # x
y1 = y.to_numpy().reshape((len(x), 1)) # x
x2 = x1.copy() # x^2
for i in range(len(x)): 
    x2[i] = x2[i] ** 2
```

```python
# Normal Equation Function
def normal_equation(x1, x2, y1):
    ones = np.ones((x.shape[0],1)) # 1D matrix of ones
    A = np.append(ones, x1, axis = 1)
    A = np.append(A, x2, axis = 1)
    W = np.dot(np.linalg.inv(np.dot(A.T, A)), np.dot(A.T, y1))
    return W

result = normal_equation(x1,x2,y1)
```

```python
# Normal Equation Plot
X = np.linspace(-3, 1, 1000)
y_guess = result[2] * X**2 + result[1] * X + result[0]

plt.scatter(x, y)
plt.plot(X, y_guess, 'red')
print(f"Result is\n{result[2]}x^2 + {result[1]}x + {result[0]}")
plt.show()
```
